## np.linspace(start, stop, num)

```python
# 0(start)에서 10(stop)사이를 11(num)개의 수로 등분
# 결국 0과 10을 포함해서 10(11-1)등분하는 수의 배열
np.linspace(0, 10, 11)
```

```python
np.linspace(0, 10, 10)
```

```python
# (stop - start) / (num - 1)
(10-0) / (10-1)
```

```python
np.linspace(0, 5, 6)
```

```python
np.linspace(0, 5, 5)
```

```python
# (stop - start) / (num - 1)
(5-0) / (5-1)
```
